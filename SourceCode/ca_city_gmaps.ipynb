```python
#import gmaps
#%matplotlib notebook
# Google developer API key
#from config import gkey
import matplotlib.pyplot as plt
import pandas as pd
import gmaps
```

```python
#setting the google key configuration
gkey = ""
# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
#reading the output file from clean_parse module
ca_city_df = pd.read_csv("../output/ca_city_gmap_input.csv")
ca_city_df.head()
```

```python
#taking one month consistant value of the year for parsing/calculations
ca_df = ca_city_df[ca_city_df["Month"] == 1].reset_index(drop = True)
ca_df.head()
#ca_df.reset_index(drop = True)
```

```python
#creating a new column combining city with median sale pricec
for i in range(len(ca_df)):
    ca_df["City Price"][i] = ca_df["City"][i] + ", $" + str(ca_df["Median Sale Price"][i])

ca_df.head()
```

```python
city_location = ca_df[["lat","lng"]]
# Customize the size of the figure
figure_layout = {
    'width': '600px',
    'height': '500px',
    'scale': 1,
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto',
    'zoom_level': 5
}
fig = gmaps.figure(layout=figure_layout)

list_of_contents = ca_df["City Price"]

# Assign the marker layer to a variable
markers = gmaps.marker_layer(city_location, info_box_content=list_of_contents)
# Add the layer to the map
fig.add_layer(markers)
fig
```
