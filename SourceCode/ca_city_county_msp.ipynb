```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
import os 
```

```python
#read the real estate data file of CA for 2012 to 2019 from Redfin
#creating data frame from the file
ca_data_df = pd.read_csv("../output/5_city_data.csv")
ca_city_data_df = ca_data_df[["City","Median Sale Price","Month","Year"]]
```

```python
#Data clean-up used 1st month of each year as start point
ca_df = ca_city_data_df[ca_city_data_df["Month"] == 1]
ca_df.reset_index(drop = True)
```

```python
#Create pivot format data year -> City -> Median sale price
ca_new_df = ca_df.pivot(index="Year", columns="City",values="Median Sale Price")
ca_new_df.head()
```

```python
#Bar charts of each city per year
ca_new_df.plot(kind="bar", align = "center")

plt.ylabel('Median Sale Price')
plt.title('Cities Median Sale Price Chart')
plt.gca().yaxis.grid(True, linestyle='--')
plt.savefig("../output/Cities Medan Sale Price.png")
plt.show()
```

```python
for col in ca_new_df:
    plt.plot(ca_new_df.index, ca_new_df[col])

plt.legend(frameon = True, title = "Cities", edgecolor = 'black')
plt.ylabel('Median Sale Price')
plt.xlabel("Year")
plt.title('Cities Median Sale Price Trend')
plt.gca().yaxis.grid(linestyle='--')
plt.savefig("../output/Cities Medan Sale Trend.png")
plt.show()
```

```python
for col in ca_new_df:
    plt.scatter(ca_new_df.index, ca_new_df[col])

plt.legend(frameon = True, title = "Cities", edgecolor = 'black')
plt.ylabel('Median Sale Price')
plt.xlabel("Year")
plt.title('Cities Median Sale Price Scatter graph')

plt.savefig("../output/Cities Medan Sale scatter.png")
plt.grid(linestyle='--')
#plt.scatter(ca_new_df.index, ca_new_df["Fremont"])
plt.show()
```
