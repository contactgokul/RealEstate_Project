```python
#import gmaps
#%matplotlib notebook
# Google developer API key
#from config import gkey
import matplotlib.pyplot as plt
import pandas as pd
import gmaps
```

```python
#setting the google key configuration
gkey = ""
# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
ca_county_df = pd.read_csv("../output/ca_county_gmap_input.csv")
ca_df = ca_county_df[ca_county_df["Month"] == 1].reset_index(drop = True)
ca_df.head()
```

```python
for i in range(len(ca_df)):
    ca_df["County Price"][i] = ca_df["County"][i] + ", $" + str(ca_df["Median Sale Price"][i])

ca_df.head()
```

```python
county_location = ca_df[["lat","lng"]]

# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto',
    'zoom_level': 5
}
fig = gmaps.figure(layout=figure_layout)

list_of_contents = ca_df["County Price"]

# Assign the marker layer to a variable
markers = gmaps.marker_layer(county_location, info_box_content=list_of_contents)
# Add the layer to the map
fig.add_layer(markers)
fig
```
